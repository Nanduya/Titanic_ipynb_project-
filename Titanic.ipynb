```python
def impute_age(col):
    Age=col[0]
    Pclass=col[1]
    if pd.isnull(Age):
        if Pclass==1:
            return df_train[df_train["Pclass"]==1]["Age"].mean()
        elif Pclass==2:
            return df_train[df_train["Pclass"]==2]["Age"].mean()
        elif Pclass==3:
            return df_train[df_train["Pclass"]==3]["Age"].mean()
    else:
        return Age
    
```

```python
df_train["Age"]=df_train[["Age","Pclass"]].apply(impute_age,axis=1)
```
